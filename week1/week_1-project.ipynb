# 문자열 매칭 알고리즘 구현 

**문제 설명**

문자열 집합 S가 존재한다 ex) S = {"www","xman","yman"}. 임의의 문자열이 N개 주어졌 을때 각 임의의 문자열 내에 집합 S에 속하는 문자열이 존재하는지 판별하시오. 아래의 문제를 순서대로 작성하세요. 1번만 푸셔도 됩니다. 

1. 문자열의 여러 부분 문자열 중 하나라도 집합 S에 있으면 'YES'를 출력하고, 아무것도 없으면 'NO'를 출력한다. 
2. 주어진 문자열의 여러부분 문자열 중 처음 매칭된 패턴과 문자열 시작 포지션을 출력한다. 
3. 주어진 문자열의 여러부분 문자열 중 매칭 가능한 모든 패턴과 문자열 시작 포지션을 출력한다. 

**입력**

- 첫째 줄에 집합 S의 크기 N이 주어진다. (1 ≤ N ≤ 1000) 
- 다음 N개 줄에 집합 S의 원소들이 주어진다. 이 문자열의 길이는 100을 넘지 않는다. 
- 다음 줄에 답을 판별해야 하는 문자열의 개수 Q가 주어진다. (1 ≤ Q ≤ 1000) 
- 다음 Q개 줄에 답을 판별해야 하는 문자열이 주어진다. 이 문자열의 길이는 10000을 넘지 않는다. 
- 입력으로 주어지는 모든 문자열은 알파벳 소문자로만 이루어져 있다. 

**출력**

Q개 줄에 각 문자열에 대한 답을 출력한다. 


"aaa",""aaaabb", ""aabbcc", "abb", "bcc", "bbcc", "aabbccdd", "aaabb", "cccd" (comma, new line 구분은 자유) 
1(매칭 대상 문자열 수) 
"aaaabbaabbccdd" 


```python
#slicing으로 풀기

n = int(input())
s = input().split(',')
q = input().split(',')

ans1 = ['NO'] * n
ans2 = {}
ans3 = []

for text in q:
    for i, pattern in enumerate(s):
        for idx in range(len(text)-len(pattern)+1):
            if text[idx:idx+len(pattern)] == pattern:
                ans1[i] = 'YES'
                ans3.append((pattern, idx))
                if not pattern in ans2:
                    ans2[pattern] = idx

print('\n1.')   
for ans in ans1:
    print(ans)
print('\n2.')
for ans in ans2:
    print(f'#pos = {ans2[ans]}, pattern = {ans}')

print('\n3.')   
for ans in ans3:
    print(f'#pos = {ans[1]}, pattern = {ans[0]}')
```

```python
#KMP알고리즘으로 풀기

def piList(pattern):
    m, j = len(pattern), 0
    pi = [0] * m
    
    for i in range(1, m):
        while j and pattern[i] != pattern[j]: 
            j = pi[j - 1]
        if pattern[i] == pattern[j]: 
            j += 1
            pi[i] = j
            
    return pi

def KMP(text, pattern):
    N = len(text)
    M = len(pattern)

    pi = piList(pattern)

    j = 0

    for i in range(N):
        if text[i] == pattern[j]:
            if j + 1 == M :
                return 'YES'
            else:
                j += 1
        else:
            while j > 0 :
                j = pi[j-1]
                try:
                    assert text[i] == pattern[j]
                except:
                    continue
                else:
                    j += 1
                    break

    return 'NO'

for text in q:
    for pattern in s:
        print(KMP(text, pattern))
```
