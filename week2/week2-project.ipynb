```python
#grouped barplot으로 시각화

def draw_group_barchart(df,main_category,sub_category,fig_width=14,fig_height=10, \
                        between_bar_padding=0.85,\
                        within_bar_padding=0.8, config_bar=None):
    

    import matplotlib.pyplot as plt
    import seaborn as sns
    import numpy as np
    %matplotlib inline 

 
    num_sub_category = len(sub_category) ## 서브 카테고리 개수

    fig = plt.figure(figsize=(fig_width,fig_height)) ## 캔버스 생성
    fig.set_facecolor('white') ## 캔버스 색상 지정
    ax = fig.add_subplot() ## 그림이 그려질 축을 생성
    
    colors = ['#00A0A9', '#47212D', '#C70000', '#FF7400', '#FFCC51'] ## 막대기 색상 지정
    
    tick_label = list(df.index) ## 메인 카테고리 라벨 생성
    tick_number = len(tick_label) ## 메인 카테고리 눈금 개수
    
    tick_coord = np.arange(tick_number) ## 메인 카테고리안에서 첫번째 서브 카테고리 막대기가 그려지는 x좌표
 
    width = 1/num_sub_category*between_bar_padding ## 막대기 폭 지정
 
    config_tick = dict()
    config_tick['ticks'] = [t + width*(num_sub_category-1)/2 for t in tick_coord] ## 메인 카테고리 라벨 x좌표
    config_tick['labels'] = tick_label 

    plt.legend() ## 범례 생성
 

    plt.yticks(**config_tick) ## x축 눈금 라벨 생성

    for i in range(num_sub_category):
        if config_bar: # 바 차트 추가 옵션이 있는 경우
            ax.barh(tick_coord+width*i, df[sub_category[i]], \
                    width*within_bar_padding, label=sub_category[i], \
                    color=colors[i], **config_bar) ## 수평 바 차트 생성
        else:
            ax.barh(tick_coord+width*i, df[sub_category[i]], \
                    width*within_bar_padding, label=sub_category[i], \
                    color=colors[i]) ## 수평 바 차트 생성
    plt.legend() ## 범례 생성
    plt.show()


main_category = df_tag_wide.index
sub_category = df_tag_wide.columns
 
 
draw_group_barchart(df_tag_wide,main_category,sub_category,\
                        between_bar_padding=0.8,\
                        within_bar_padding=1, config_bar = None)
```

# 2. 각각의 비디오는 시청자의 호응도(engagement)를 판단할수 있는 객관적인 지표들이 있음

ex) views, likes, dislikes, comments,...


*   비디오를 인기 동영상 기준에 부합하도록 분류할수 있는 새로운 지표를 개발하고
*   이 지표를 사용하여 engagement 와 어떤 상관관계가 있는지 설명하시오.




```python
df.columns
```

```python
df_en = df.copy()
df_en['trending_date_term'] = pd.to_datetime(df_en['off_trending_date']) - pd.to_datetime(df_en['on_trending_date'])
df_en[['trending_date_term']].sort_values(by = 'trending_date_term', ascending = 0)
```

- 인기 동영상 기준에 부합하는 동영상일수록 인기동영상을 유지하는 기간이 길 것
- 데이터를 인기동영상을 유지한 기간으로 나누어 분석

```python
# 인기동영상으로 올라가 있는 기간 동안 증가한 수치를 나타는 column 추가
df_en['views'] = df_en['off_views'] - df_en['on_views']
df_en['likes'] = df_en['off_likes'] - df_en['on_likes']
df_en['dislikes'] = df_en['off_dislikes'] - df_en['on_dislikes']
df_en['comments'] = df_en['off_comments'] - df_en['on_comments']
df_en['view_differ'] =  df_en['off_views'] - df_en['on_views']

df_en[['video_id', 'channel_id', 'duration', 'trending_date_term',	'views',
       'likes',	'dislikes',	'comments', 'on_channel_total_views', 'view_differ']].sort_values(by = 'trending_date_term', ascending = 0)
```

```python
# duration 단위를 분으로 통일
def toMinutes(string):
    hour, minute, second = 0, 0, 0

    string = string.replace('PT', '')
    if 'H' in string:
        hour, string = int(string.split('H')[0]), string.split('H')[1]
    if 'M' in string:
        minute, string = int(string.split('M')[0]), string.split('M')[1]
    if 'S' in string:
        second = int(string.replace('S', ''))

    return (60 * hour) + minute + second / 60
```

```python
df_en['duration'] = df_en['duration'].apply(toMinutes)
```

```python
df_en = df_en[['video_id', 'channel_id', 'duration', 'trending_date_term', 'views', 
               'likes', 'dislikes', 'comments', 'on_channel_total_views', 'view_differ']]
df_en.describe(include='all')
```

```python
# 좋아요수 대비 싫어요 수의 비율 column 추가
df_en['dislike_ratio'] = (df_en['dislikes'] / df_en['likes']) * 100
# 좋아요수 대비 코멘트 수의 비율 column 추가
df_en['comment_ratio'] = (df_en['comments'] / df_en['likes']) * 100
# 조회수 증가율 column 추가
df_en['view_ratio'] = (df_en['view_differ'] / df_en['on_channel_total_views']) * 100

df_en[['dislike_ratio', 'comment_ratio']].fillna(0, inplace = True)
```

```python
# 인기 급상승 동영상에 머무른 기간으로 채널을 분류하여 분석
df_en_0 = df_en[df_en['trending_date_term'] == '0 days']
```

```python
# 기간별 동영상 길이 비교
df_en_dura = df_en_0.describe()['duration']

# 기간 6와 7은 데이터의 수가 적어서 빼고 확인
for i in range(1, 6):
    df_en_dura = pd.concat([df_en_dura, df_en[df_en['trending_date_term'] == f'{i} days'].describe()['duration']], axis = 1)

df_en_dura = df_en_dura.T
df_en_dura.index = [f'{i} days' for i in range(6)]
df_en_dura = df_en_dura[['25%', '50%', '75%', 'mean']]
# df_en_dura['mean'] = (df_en_dura['25%'] + df_en_dura['50%'] + df_en_dura['75%']) / 3
df_en_dura
```

```python
#line plot 으로 시각화
import matplotlib.pyplot as plt

plt.figure(figsize=(13,6)) 


plt.xlabel('trending_date_term')           
plt.ylabel('duration')


plt.plot(df_en_dura['25%'], label='25%', c='#99A345')
plt.plot(df_en_dura['50%'], label='50%', c='#5AABA4')
plt.plot(df_en_dura['75%'], label='75%', c='#9F90ED')
plt.plot(df_en_dura['mean'], label='mean', c='#E6798A')



plt.legend(loc=1)

plt.show()
```

- 동영상의 길이는 인기 동영상 유지기간이 변함에 따라 그 수치가 일관성을 보이지 않는다.
